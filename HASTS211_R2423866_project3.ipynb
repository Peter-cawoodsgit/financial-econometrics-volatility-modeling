<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">1. Definition</span></div>

<div style="border-bottom:1px solid rgba(247,148,29,0.3); padding:6px 4px 6px 12px; margin:14px 0 4px 0;"><span style="font-size:1em; font-weight:600; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif;">1.1 Non-Stationarity and Unit Roots</span></div>

A time series $y_t$ is said to be **integrated of order $d$**, written $y_t \sim I(d)$, if it must be differenced $d$ times to achieve covariance-stationarity. Formally, $y_t$ is *non-stationary* if its mean or variance changes over time:

$$
y_t = \mu + y_{t-1} + \varepsilon_t, \quad \varepsilon_t \overset{iid}{\sim} (0,\sigma^2)
$$

This is the **Random Walk** (unit root) model, where shocks $\varepsilon_t$ accumulate permanently.

<div style="border-bottom:1px solid rgba(247,148,29,0.3); padding:6px 4px 6px 12px; margin:14px 0 4px 0;"><span style="font-size:1em; font-weight:600; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif;">1.2 Cointegration</span></div>

Let $\mathbf{y}_t = (y_{1t},\, y_{2t},\ldots,y_{kt})^\top$ be a $k\times1$ vector of $I(1)$ series. The system is **cointegrated** of rank $r$ if there exists a $k\times r$ matrix $\boldsymbol{\beta}$ with full column rank such that:

$$
\boldsymbol{\beta}^\top \mathbf{y}_t \sim I(0)
$$

The columns of $\boldsymbol{\beta}$ are the **cointegrating vectors** — linear combinations that strip out the common stochastic trend and yield a stationary *equilibrium error*.

<div style="border-bottom:1px solid rgba(247,148,29,0.3); padding:6px 4px 6px 12px; margin:14px 0 4px 0;"><span style="font-size:1em; font-weight:600; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif;">1.3 Vector Error Correction Model (VECM)</span></div>

For a cointegrated system with $r$ cointegrating relationships and $p-1$ lags, the VECM is:

$$
\Delta\mathbf{y}_t = \boldsymbol{\alpha}\boldsymbol{\beta}^\top \mathbf{y}_{t-1} + \sum_{i=1}^{p-1}\boldsymbol{\Gamma}_i\,\Delta\mathbf{y}_{t-i} + \boldsymbol{\mu} + \boldsymbol{\varepsilon}_t
$$

where:

| Symbol | Definition |
|---|---|
| $\Delta\mathbf{y}_t$ | First difference of the price vector (short-run changes) |
| $\boldsymbol{\alpha}$ | $k\times r$ **adjustment/loading matrix** — speed of mean-reversion |
| $\boldsymbol{\beta}$ | $k\times r$ **cointegrating matrix** — long-run equilibrium relationships |
| $\boldsymbol{\beta}^\top \mathbf{y}_{t-1}$ | *Error-correction term* (ECT): deviation from long-run equilibrium |
| $\boldsymbol{\Gamma}_i$ | $k\times k$ short-run lag coefficient matrices |
| $\boldsymbol{\mu}$ | Intercept vector |
| $\boldsymbol{\varepsilon}_t$ | White-noise disturbance: $E[\boldsymbol{\varepsilon}_t]=\mathbf{0}$, $E[\boldsymbol{\varepsilon}_t\boldsymbol{\varepsilon}_t^\top]=\boldsymbol{\Omega}$ |

<div style="border-bottom:1px solid rgba(247,148,29,0.3); padding:6px 4px 6px 12px; margin:14px 0 4px 0;"><span style="font-size:1em; font-weight:600; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif;">1.4 Johansen Trace and Maximum-Eigenvalue Statistics</span></div>

The Johansen (1988) procedure tests $H_0: r \leq r_0$ using:

$$
\lambda_{\text{trace}}(r_0) = -T \sum_{i=r_0+1}^{k} \ln(1-\hat{\lambda}_i)
$$

$$
\lambda_{\max}(r_0, r_0+1) = -T\ln(1-\hat{\lambda}_{r_0+1})
$$

where $\hat{\lambda}_i$ are the ordered eigenvalues of the reduced-rank regression.


---
<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">2. Description</span></div>

**Cointegration** is a long-run equilibrium concept: even though two or more financial price series individually wander without a fixed mean (i.e., are non-stationary), a particular linear combination of them is stationary, meaning they are perpetually pulled back toward each other by market forces.

The **Vector Error Correction Model (VECM)** decomposes the dynamics of a cointegrated system into two parts: a long-run equilibrium component captured by the error-correction term (ECT), and short-run adjustment dynamics captured by lagged differences, with the loading matrix $\boldsymbol{\alpha}$ governing how quickly each variable reverts toward equilibrium after a shock.

<div style="border-bottom:1px solid rgba(247,148,29,0.3); padding:6px 4px 6px 12px; margin:14px 0 4px 0;"><span style="font-size:1em; font-weight:600; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif;">Why This Dataset Works Well</span></div>

Apple Inc. (AAPL), Microsoft Corp. (MSFT), and the SPDR Gold ETF (GLD) are suitable for cointegration analysis for the following reasons:

- **AAPL and MSFT** are both mega-cap technology stocks subject to the same macro risk factors (interest rates, tech sector sentiment, broad market cycles), creating a plausible shared stochastic trend.
- **GLD** serves as an anchor asset: its inclusion tests whether equity markets are cointegrated with commodity safe-haven assets over the long run.
- All three series exhibit **price-level non-stationarity** (confirmed by ADF tests below), making them candidates for $I(1)$ classification.
- The 10-year daily window (2015–2024) spans multiple regimes (bull markets, COVID crash, post-pandemic rally, rate hike cycle), providing statistical power for cointegration detection.
- This is **not** a simple unit root test on a single series (as done in GWP2); instead, it tests for multivariate long-run relationships using the Johansen procedure and estimates a full VECM.

```python
# ── 3.1 Download Data from Yahoo Finance ──────────────────────────────────────
TICKERS = ['AAPL', 'MSFT', 'GLD']
START   = '2015-01-01'
END     = '2024-12-31'

raw = yf.download(TICKERS, start=START, end=END, auto_adjust=True, progress=True)
prices = raw['Close'].dropna()
prices.columns.name = None

print(f'\n── Dataset Summary ──')
print(f'Tickers   : {TICKERS}')
print(f'Frequency : Daily')
print(f'Period    : {prices.index[0].date()} → {prices.index[-1].date()}')
print(f'Total obs : {len(prices):,}')
print(f'Source    : Yahoo Finance (via yfinance library)')
print(f'Units     : USD (adjusted closing price)\n')
print(prices.head(5))
print('...')
print(prices.tail(5))
```

<div style="border-left:5px solid #F7941D; background:#0B0C2A; border-radius:0 6px 6px 0; padding:10px 20px; margin:18px 0 6px 0;"><span style="font-size:1.1em; font-weight:700; color:#F7941D; font-family:'Segoe UI',Arial,sans-serif; letter-spacing:1px;">4. Demonstration — Model Estimation & Output</span></div>

The demonstration follows these steps:
1. ADF unit root tests (confirm $I(1)$)
2. KPSS stationarity tests (cross-confirmation)
3. Johansen cointegration test (determine rank $r$)
4. VAR lag selection (information criteria)
5. VECM estimation (fit the model)
6. Parameter interpretation

```python
# ── 4.1 Augmented Dickey-Fuller Tests ──────────────────────────────────────
# H₀: Unit root present (series is non-stationary)
# Reject H₀ if p-value < 0.05

def adf_summary(series, name, max_lag=20):
    result = adfuller(series.dropna(), maxlag=max_lag, autolag='AIC')
    return {
        'Series': name,
        'ADF Statistic': round(result[0], 4),
        'p-value': round(result[1], 4),
        'Lags Used': result[2],
        'Obs': result[3],
        '1% CV': round(result[4]['1%'], 4),
        '5% CV': round(result[4]['5%'], 4),
        '10% CV': round(result[4]['10%'], 4),
        'Stationary?': 'YES' if result[1] < 0.05 else 'NO'
    }

print('=== ADF Test — LEVEL PRICES (H₀: Unit root present) ===')
level_results = []
for col in TICKERS:
    level_results.append(adf_summary(prices[col], col))
adf_levels = pd.DataFrame(level_results).set_index('Series')
print(adf_levels.to_string())

print('\n=== ADF Test — FIRST DIFFERENCES (H₀: Unit root present) ===')
diff_results = []
for col in TICKERS:
    diff_results.append(adf_summary(prices[col].diff().dropna(), f'Δ{col}'))
adf_diffs = pd.DataFrame(diff_results).set_index('Series')
print(adf_diffs.to_string())
```

```python
# ── 4.2 KPSS Tests (cross-confirmation of non-stationarity) ─────────────────
# H₀: Series IS stationary → reject H₀ if p-value < 0.05

def kpss_summary(series, name):
    stat, pval, lags, crit = kpss(series.dropna(), regression='ct', nlags='auto')
    return {
        'Series': name,
        'KPSS Statistic': round(stat, 4),
        'p-value': round(pval, 4) if pval != 0.01 else '<0.01',
        '5% CV': round(crit['5%'], 4),
        'Non-Stationary?': 'YES' if pval < 0.05 else 'NO'
    }

print('=== KPSS Test — LEVEL PRICES (H₀: Series is stationary) ===')
kpss_rows = [kpss_summary(prices[col], col) for col in TICKERS]
print(pd.DataFrame(kpss_rows).set_index('Series').to_string())
```

```python
# ── 4.3 Johansen Cointegration Test ──────────────────────────────────────────
# Tests for the number of cointegrating vectors r
# det_order=0 → constant in the cointegrating equation
# k_ar_diff=4 → lags in the VECM (to be refined by information criteria)

joh_result = coint_johansen(prices[TICKERS], det_order=0, k_ar_diff=4)

print('=== Johansen Cointegration Test ===')
print(f'Hypothesis: H₀: rank ≤ r  vs  H₁: rank > r')
print(f'Number of series (k) = {len(TICKERS)}')
print()

# Trace statistic table
print('--- TRACE Statistic ---')
trace_df = pd.DataFrame({
    'H₀: r ≤': list(range(len(TICKERS))),
    'Trace Stat': joh_result.lr1.round(4),
    'CV 90%': joh_result.cvt[:, 0].round(4),
    'CV 95%': joh_result.cvt[:, 1].round(4),
    'CV 99%': joh_result.cvt[:, 2].round(4),
    'Reject H₀ (5%)?': ['YES' if joh_result.lr1[i] > joh_result.cvt[i, 1] else 'NO'
                         for i in range(len(TICKERS))]
})
print(trace_df.to_string(index=False))

print()
print('--- MAX-EIGENVALUE Statistic ---')
maxeig_df = pd.DataFrame({
    'H₀: r =': list(range(len(TICKERS))),
    'Max-Eig Stat': joh_result.lr2.round(4),
    'CV 90%': joh_result.cvm[:, 0].round(4),
    'CV 95%': joh_result.cvm[:, 1].round(4),
    'CV 99%': joh_result.cvm[:, 2].round(4),
    'Reject H₀ (5%)?': ['YES' if joh_result.lr2[i] > joh_result.cvm[i, 1] else 'NO'
                         for i in range(len(TICKERS))]
})
print(maxeig_df.to_string(index=False))

print()
print('--- Normalised Cointegrating Vectors (β) ---')
beta_df = pd.DataFrame(joh_result.evec.T,
                       columns=TICKERS,
                       index=[f'CV {i+1}' for i in range(len(TICKERS))])
print(beta_df.round(6).to_string())
```

```python
# ── 4.4 VAR Lag Order Selection (for the VECM) ─────────────────────────────
# The VECM with p-1 lags corresponds to a VAR(p) in levels

print('=== VAR Lag Order Selection (Information Criteria) ===')

var_model = VAR(prices[TICKERS])
lag_order_result = var_model.select_order(maxlags=10)
print(lag_order_result.summary())

# Extract optimal lag
optimal_lag_aic = lag_order_result.aic
optimal_lag_bic = lag_order_result.bic
print(f'\nAIC selects: {optimal_lag_aic} lags')
print(f'BIC selects: {optimal_lag_bic} lags')

# Use BIC (more parsimonious) for VECM
chosen_lag = max(optimal_lag_bic - 1, 1)  # VECM uses p-1 lags
print(f'\nChosen lags for VECM (BIC - 1): {chosen_lag}')
```

```python
# ── 4.5 Select Cointegration Rank ───────────────────────────────────────────
rank_result = select_coint_rank(prices[TICKERS], det_order=0,
                                 k_ar_diff=chosen_lag, signif=0.05)
r = rank_result.rank
print(f'=== Selected Cointegration Rank ===')
print(f'Johansen test selects r = {r} cointegrating relationship(s) at 5% significance.')
```

```python
# ── 4.6 Estimate VECM ──────────────────────────────────────────────────────
# Fit the Vector Error Correction Model

vecm_fit = VECM(prices[TICKERS],
                k_ar_diff=chosen_lag,
                coint_rank=r,
                deterministic='ci').fit()

print(vecm_fit.summary())
```

```python
# ── 4.7 Parameter Interpretation ──────────────────────────────────────────

print('='*65)
print('CALIBRATED PARAMETER INTERPRETATION')
print('='*65)

print('\n[A] COINTEGRATING VECTOR (β) — Long-Run Equilibrium')
print('-'*65)
beta = vecm_fit.beta
beta_df = pd.DataFrame(beta, index=TICKERS,
                        columns=[f'CV {i+1}' for i in range(r)])
print(beta_df.round(6).to_string())
print('\nInterpretation: β defines the long-run equilibrium constraint.')
print('  The ECT = β\'y_{t-1} is stationary despite individual series being I(1).')
print('  Normalised on AAPL: a 1-unit rise in AAPL corresponds to a βᴹˢᶠᵀ-unit')
print('  move in MSFT to maintain equilibrium, and similarly for GLD.')

print('\n[B] ADJUSTMENT MATRIX (α) — Speed of Mean Reversion')
print('-'*65)
alpha = vecm_fit.alpha
alpha_df = pd.DataFrame(alpha, index=TICKERS,
                         columns=[f'α (ECT {i+1})' for i in range(r)])
print(alpha_df.round(6).to_string())
print()
for i, ticker in enumerate(TICKERS):
    a_val = alpha[i, 0]
    direction = 'corrects toward' if a_val < 0 else 'moves away from'
    speed = abs(a_val) * 100
    print(f'  {ticker}: α = {a_val:.6f} → {ticker} {direction} equilibrium.')
    print(f'         Each period, {ticker} closes {speed:.3f}% of the disequilibrium gap.')

print()
print('[C] SHORT-RUN COEFFICIENTS (Γᵢ) — Lag Dynamics')
print('-'*65)
print('  The Γᵢ matrices capture cross-variable short-run dynamics.')
print('  Significant Γᵢ coefficients indicate Granger causality in the short run.')
print('  (See the VECM summary output above for individual t-statistics.)')
```

```python
# ── 6.2 Ljung-Box Test for Residual Autocorrelation ───────────────────────
print('=== Ljung-Box Test on VECM Residuals (H₀: No autocorrelation) ===')
print(f'{"Series":<8} {"Lag 5":<20} {"Lag 10":<20} {"Lag 20":<20}')
print('-'*68)
for col in TICKERS:
    lb5  = acorr_ljungbox(residuals[col], lags=[5],  return_df=True)
    lb10 = acorr_ljungbox(residuals[col], lags=[10], return_df=True)
    lb20 = acorr_ljungbox(residuals[col], lags=[20], return_df=True)
    p5   = lb5['lb_pvalue'].values[0]
    p10  = lb10['lb_pvalue'].values[0]
    p20  = lb20['lb_pvalue'].values[0]
    print(f'{col:<8} p={p5:.4f} {"✓" if p5>0.05 else "✗"}         p={p10:.4f} {"✓" if p10>0.05 else "✗"}         p={p20:.4f} {"✓" if p20>0.05 else "✗"}')

print()
print('✓ = Fail to reject H₀ (residuals appear uncorrelated) — desirable')
print('✗ = Reject H₀ (residual autocorrelation detected) — problematic')
```

```python
# ── 6.3 QQ-Plots & Normality Tests ──────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(14, 5))
for ax, col, color in zip(axes, TICKERS, colors):
    stats.probplot(residuals[col], dist='norm', plot=ax)
    ax.set_title(f'Q-Q Plot — {col}', fontsize=10)
    ax.get_lines()[0].set(color=color, markersize=2, alpha=0.5)
    ax.get_lines()[1].set(color='red', linewidth=1.5)
    ax.grid(True, alpha=0.3)

fig.suptitle('Figure 9: Normal Q-Q Plots of VECM Residuals', fontsize=12)
plt.tight_layout()
plt.savefig('fig9_qq.png', dpi=130, bbox_inches='tight')
plt.show()

print('=== Jarque-Bera Normality Tests on Residuals ===')
for col in TICKERS:
    stat, pval = stats.jarque_bera(residuals[col])
    print(f'{col}: JB stat = {stat:.4f}, p-value = {pval:.6f} '
          f'→ {"Normal" if pval > 0.05 else "Non-Normal"} at 5%')
```

## DAMAGE ASSESSMENT — Model Problems & Data Quality Issues

**1. NON-NORMALITY OF RESIDUALS (Jarque-Bera test)**

Financial return residuals exhibit fat tails and excess kurtosis. The VECM assumes Gaussian disturbances; violations inflate standard errors and invalidate inference on α and β. This is a near-universal challenge in equity data.
> Challenge match: *Non-Gaussian innovations / Fat tails*

---

**2. HETEROSKEDASTICITY / ARCH EFFECTS (Squared residuals)**

Volatility clustering is visually evident in squared residuals. The VECM assumes homoskedastic errors; conditional heteroskedasticity causes OLS coefficient estimates to remain consistent but inefficient, and confidence intervals to be unreliable.
> Challenge match: *Volatility clustering / Conditional heteroskedasticity*

---

**3. STRUCTURAL BREAKS (COVID-19 crash, 2022 rate-hike cycle)**

The 2020 COVID crash and the 2022 Fed tightening cycle likely altered the cointegrating relationship. A fixed-parameter VECM cannot accommodate such breaks; the α and β estimated over the full sample may be averages of fundamentally different regimes.
> Challenge match: *Parameter instability / Regime changes*

---

**4. RESIDUAL AUTOCORRELATION**

If the Ljung-Box test flags autocorrelation, the lag order p is insufficient. Remaining serial correlation inflates Type I error in hypothesis tests on the model.
> Challenge match: *Model misspecification / Insufficient lag order*

---

**5. SPURIOUS COINTEGRATION RISK**

GLD's price is driven primarily by gold commodity dynamics and USD strength, not technology earnings. Including GLD introduces risk of spurious long-run relationship detection, especially over a long window with regime changes.
> Challenge match: *Spurious regression / Overfitting cointegrating vectors*

---

**6. LOOK-AHEAD BIAS IN FORECASTING**

The full-sample β and α are estimated using all available data. A live deployment would require rolling estimation to avoid using future information.
> Challenge match: *Data snooping / Look-ahead bias*

```python
# ── 7.2 Visual: Rolling Correlation (Parameter Instability) ─────────────────
window = 252  # 1 trading year
roll_corr_am = prices['AAPL'].rolling(window).corr(prices['MSFT'])
roll_corr_ag = prices['AAPL'].rolling(window).corr(prices['GLD'])

fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(roll_corr_am.index, roll_corr_am, label='AAPL-MSFT', color='#1f77b4', linewidth=1.2)
ax.plot(roll_corr_ag.index, roll_corr_ag, label='AAPL-GLD', color='#2ca02c', linewidth=1.2)
ax.axhline(0, color='black', linewidth=0.6, linestyle='--')

# Shade COVID and rate-hike periods
ax.axvspan(pd.Timestamp('2020-02-01'), pd.Timestamp('2020-06-01'),
           alpha=0.2, color='red', label='COVID crash')
ax.axvspan(pd.Timestamp('2022-01-01'), pd.Timestamp('2022-12-31'),
           alpha=0.2, color='orange', label='Rate-hike cycle')

ax.set_ylabel('Rolling Correlation (1-year)', fontsize=10)
ax.set_xlabel('Date', fontsize=10)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)
ax.set_title('Figure 12: Rolling 1-Year Correlation — Parameter Instability Across Regimes', fontsize=11)
plt.tight_layout()
plt.savefig('fig12_rolling_corr.png', dpi=130, bbox_inches='tight')
plt.show()
print('Figure 12: The correlation between AAPL-MSFT fluctuates across regimes, challenging a fixed VECM.')
```

### Direction 3: Handling Heteroskedasticity — VECM-GARCH

To address volatility clustering in VECM residuals, a multivariate GARCH model (e.g., DCC-GARCH or BEKK-GARCH) can be grafted onto the VECM:

- **Step 1:** Estimate VECM → extract residuals ε_t
- **Step 2:** Model the conditional covariance of ε_t via:
  -    *(BEKK)*
- **Step 3:** Re-estimate VECM with QMLE using H_t weights

This approach produces:
- More efficient parameter estimates
- Correct standard errors despite heteroskedasticity
- Time-varying conditional correlations (DCC variant)

**Implementation:** Python's x86_64 library or R's  package.

**Reference:** Engle & Granger (1987); Johansen (1988); Engle (2002) DCC.

## DEPLOYMENT BLUEPRINT — VECM TRADING SYSTEM

**1. DATA PIPELINE**
- Pull daily adjusted close prices via  or a Bloomberg data feed each evening at market close.
- Maintain a rolling 2-year history for VECM re-estimation.

---

**2. MODEL RE-ESTIMATION (Weekly)**
- Re-run Johansen test and VECM every week on the rolling sample to capture shifts in cointegration rank and α values.
- Flag if rank r changes → alert risk manager.

---

**3. SIGNAL GENERATION (Daily)**
- Compute ECT = β'y_t at each day's close.
- Compute Z-score of ECT using rolling 252-day mean and std.
- **ENTRY:** Z > +1.5 → short spread (sell AAPL, buy MSFT); Z < −1.5 → long spread (buy AAPL, sell MSFT)
- **EXIT:** Z reverts within ±0.3 of mean, or after 10 days.

---

**4. POSITION SIZING**
- Use |α| as weights: faster-adjusting asset gets larger weight.
- Cap gross notional at 2× portfolio NAV.

---

**5. RISK MANAGEMENT**
- Stop-loss if ECT Z-score exceeds ±3σ (cointegration break).
- Monitor rolling correlation; suspend if AAPL-MSFT corr < 0.6.
- Drawdown limit: halt trading if monthly PnL < −2%.

---

**6. REPORTING**
- Daily dashboard: ECT level, Z-score, open positions, PnL.
- Weekly report: α stability, Johansen rank, residual ACF.
